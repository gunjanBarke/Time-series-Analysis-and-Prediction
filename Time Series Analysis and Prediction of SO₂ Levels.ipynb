```python
# ‚úÖ FINAL NOTEBOOK: Modeling + Transferability

# üì¶ Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# üìÇ Dataset Loader + Cleaner
def load_dataset(filepath):
    if filepath.endswith('.xlsx'):
        df = pd.read_excel(filepath)
    elif filepath.endswith('.csv'):
        df = pd.read_csv(filepath)
    else:
        raise ValueError("Unsupported file format.")

    df.columns = df.columns.str.strip()

    # Rename for consistency
    if 'Ozone' in df.columns:
        df.rename(columns={'Ozone': 'O3'}, inplace=True)
    for col in ['Date', 'DATE', 'Datetime', 'timestamp']:
        if col in df.columns:
            df.rename(columns={col: 'date'}, inplace=True)
    return df

# ü§ñ Modeling Function
def build_and_evaluate_model(df, feature_cols, target, city_name):
    # Keep only features that actually exist in the dataset
    available_features = [col for col in feature_cols if col in df.columns]

    if target not in df.columns:
        raise ValueError(f"‚ùå Target '{target}' not found in dataset: {city_name}")

    print(f"\nüìå {city_name} - Using features: {available_features}")

    df = df.dropna(subset=[target] + available_features)
    df = df.sort_values('date') if 'date' in df.columns else df


    X = df[available_features]
    y = df[target]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    split = int(len(X) * 0.8)
    X_train, X_test = X_scaled[:split], X_scaled[split:]
    y_train, y_test = y.iloc[:split], y.iloc[split:]

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f"‚úÖ {city_name} - RMSE for {target}: {rmse:.2f}")

    # Plot predictions
    plt.figure(figsize=(10, 5))
    plt.plot(np.arange(len(y_test)), y_test.values, label='Actual')
    plt.plot(np.arange(len(y_test)), y_pred, label='Predicted')
    plt.title(f'{city_name} - {target} Prediction')
    plt.legend()
    plt.show()

    # üìä Feature Importance Plot
    importances = model.feature_importances_
    feature_names = available_features
    importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 5))
    sns.barplot(data=importance_df, x='Importance', y='Feature')
    plt.title(f'{city_name} - Feature Importance for {target}')
    plt.xlabel('Importance')
    plt.ylabel('Feature')
    plt.tight_layout()
    plt.show()

    # Export results
    result_df = pd.DataFrame({'Actual': y_test.values, 'Predicted': y_pred})
    result_df.to_csv(f'{city_name}_SO2_results.csv', index=False)

    return model, rmse

# üéØ Features & Target
feature_cols = ['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'Temp', 'RH', 'WS']
target = 'SO2'

# üîÅ Train on Gaziabad
df_gaziabad = load_dataset('/content/Gaziabad_knn_imputed_data.xlsx')
model, rmse_gz = build_and_evaluate_model(df_gaziabad, feature_cols, target, 'Gaziabad')

# üîÅ Test on Huairou
df_huairou = load_dataset('/content/Huairou_IIET.csv')
_, rmse_hr = build_and_evaluate_model(df_huairou, feature_cols, target, 'Huairou')

# üìä Summary
summary = pd.DataFrame({
    'City': ['Gaziabad', 'Huairou'],
    'RMSE': [rmse_gz, rmse_hr]
})
summary.to_csv('Model_Comparison_Summary.csv', index=False)
print("\nüìÅ Results exported:")
print(summary)
```

```python
# üì¶ Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# üì¶ Dataset Loader

def load_dataset(filepath):
    if filepath.endswith('.xlsx'):
        df = pd.read_excel(filepath)
    elif filepath.endswith('.csv'):
        df = pd.read_csv(filepath)
    else:
        raise ValueError("Unsupported file format.")

    df.columns = df.columns.str.strip()

    if 'Ozone' in df.columns:
        df.rename(columns={'Ozone': 'O3'}, inplace=True)

    for col in ['Date', 'DATE', 'Datetime', 'timestamp']:
        if col in df.columns:
            df.rename(columns={col: 'date'}, inplace=True)

    return df


# ü§ñ Enhanced Modeling Function

def build_and_evaluate_model(df, feature_cols, target, city_name):
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        df['month'] = df['date'].dt.month
        df['dayofweek'] = df['date'].dt.dayofweek

    for col in ['SO2', 'PM2.5', 'PM10']:
        if col in df.columns:
            df[f'{col}_lag1'] = df[col].shift(1)
            df[f'{col}_roll3'] = df[col].rolling(window=3).mean()

    df = df.dropna()

    engineered_features = feature_cols + ['month', 'dayofweek']
    for col in ['SO2_lag1', 'PM2.5_lag1', 'PM2.5_roll3', 'PM10_lag1', 'PM10_roll3']:
        if col in df.columns:
            engineered_features.append(col)

    available_features = [col for col in engineered_features if col in df.columns]
    if target not in df.columns:
        raise ValueError(f"Target {target} not found in dataset: {city_name}")

    print(f"\nüßπ {city_name} - Using features: {available_features}")

    X = df[available_features]
    y = df[target]

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    split = int(len(X) * 0.8)
    X_train, X_test = X_scaled[:split], X_scaled[split:]
    y_train, y_test = y.iloc[:split], y.iloc[split:]

    param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'max_features': ['sqrt', 'log2', None]  # fixed!
    }

    grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid,
                               cv=3, scoring='neg_root_mean_squared_error',
                               verbose=0, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_

    y_pred = best_model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f"üßπ {city_name} - RMSE for {target}: {rmse:.2f}")
    print(f"üîß Best Parameters: {grid_search.best_params_}")

    plt.figure(figsize=(10, 5))
    plt.plot(y_test.values, label='Actual')
    plt.plot(y_pred, label='Predicted')
    plt.title(f'{city_name} - {target} Prediction')
    plt.legend()
    plt.show()

    importances = best_model.feature_importances_
    feature_names = available_features
    importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 5))
    sns.barplot(data=importance_df, x='Importance', y='Feature')
    plt.title(f'{city_name} - Feature Importance for {target}')
    plt.tight_layout()
    plt.show()

    return best_model, rmse

# üì¶ Example Usage

# Example for Gaziabad
df_gz = load_dataset('/content/Gaziabad_knn_imputed_data.xlsx')
feature_cols = ['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'Temp', 'RH', 'WS']
target = 'SO2'

model_gz, rmse_gz = build_and_evaluate_model(df_gz, feature_cols, target, 'Gaziabad')

# Example for Huairou
df_hr = load_dataset('/content/Huairou_IIET.csv')
model_hr, rmse_hr = build_and_evaluate_model(df_hr, feature_cols, target, 'Huairou')
```

```python
# ‚úÖ FINAL NOTEBOOK: Modeling + Transferability (Improved)

# üì¶ Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

# üìÇ Dataset Loader + Cleaner
def load_dataset(filepath):
    if filepath.endswith('.xlsx'):
        df = pd.read_excel(filepath)
    elif filepath.endswith('.csv'):
        df = pd.read_csv(filepath)
    else:
        raise ValueError("Unsupported file format.")

    df.columns = df.columns.str.strip()
    if 'Ozone' in df.columns:
        df.rename(columns={'Ozone': 'O3'}, inplace=True)
    for col in ['Date', 'DATE', 'Datetime', 'timestamp']:
        if col in df.columns:
            df.rename(columns={col: 'date'}, inplace=True)
    return df

# ü§ñ Enhanced Modeling Function with Improvements
def build_and_evaluate_model(df, feature_cols, target, city_name):
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        df['month'] = df['date'].dt.month
        df['dayofweek'] = df['date'].dt.dayofweek

    # Feature engineering
    for col in ['SO2', 'PM2.5', 'PM10']:
        if col in df.columns:
            df[f'{col}_lag1'] = df[col].shift(1)
            df[f'{col}_lag2'] = df[col].shift(2)
            df[f'{col}_lag3'] = df[col].shift(3)
            df[f'{col}_roll3'] = df[col].rolling(window=3).mean()
            df[f'{col}_std3'] = df[col].rolling(window=3).std()

    df = df.dropna()

    engineered_features = feature_cols + ['month', 'dayofweek']
    for col in df.columns:
        if any(key in col for key in ['lag', 'roll', 'std']):
            engineered_features.append(col)

    available_features = [col for col in engineered_features if col in df.columns]
    if target not in df.columns:
        raise ValueError(f"Target {target} not found in dataset: {city_name}")

    print(f"\nüßπ {city_name} - Using features: {available_features}")

    # Feature Distribution
    plt.figure(figsize=(15, 10))
    df[available_features + [target]].hist(bins=30, figsize=(15, 10))
    plt.suptitle(f'{city_name} - Feature Distributions Before Training', fontsize=16)
    plt.tight_layout()
    plt.show()

    # Modeling
    X = df[available_features]
    y = df[target]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    split = int(len(X) * 0.8)
    X_train, X_test = X_scaled[:split], X_scaled[split:]
    y_train, y_test = y.iloc[:split], y.iloc[split:]

    tscv = TimeSeriesSplit(n_splits=3)
    param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [10, 20, None],
        'min_samples_split': [2, 5],
        'max_features': ['sqrt', 'log2', None]
    }

    grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid,
                               cv=tscv, scoring='neg_root_mean_squared_error',
                               verbose=0, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_

    y_pred = best_model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    print(f"üßπ {city_name} - RMSE for {target}: {rmse:.2f}")
    print(f"üîß Best Parameters: {grid_search.best_params_}")

    # Prediction Plot
    plt.figure(figsize=(10, 5))
    plt.plot(y_test.values, label='Actual')
    plt.plot(y_pred, label='Predicted')
    plt.title(f'{city_name} - {target} Prediction')
    plt.legend()
    plt.show()

    # Feature Importance
    importances = best_model.feature_importances_
    importance_df = pd.DataFrame({'Feature': available_features, 'Importance': importances})
    importance_df = importance_df.sort_values(by='Importance', ascending=False)

    plt.figure(figsize=(10, 5))
    sns.barplot(data=importance_df, x='Importance', y='Feature')
    plt.title(f'{city_name} - Feature Importance for {target}')
    plt.tight_layout()
    plt.show()

    return best_model, rmse

# üéØ Features & Target
feature_cols = ['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'Temp', 'RH', 'WS']
target = 'SO2'

# Run for Gaziabad
df_gz = load_dataset('/content/Gaziabad_knn_imputed_data.xlsx')
model_gz, rmse_gz = build_and_evaluate_model(df_gz, feature_cols, target, 'Gaziabad')

# Run for Huairou
df_hr = load_dataset('/content/Huairou_IIET.csv')
model_hr, rmse_hr = build_and_evaluate_model(df_hr, feature_cols, target, 'Huairou')
```

```python
# ‚úÖ FINAL NOTEBOOK: Modeling + Transferability (Improved with XGBoost & SHAP)

# üì¶ Required Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_percentage_error
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
import shap

# üìÇ Dataset Loader + Cleaner
def load_dataset(filepath):
    if filepath.endswith('.xlsx'):
        df = pd.read_excel(filepath)
    elif filepath.endswith('.csv'):
        df = pd.read_csv(filepath)
    else:
        raise ValueError("Unsupported file format.")

    df.columns = df.columns.str.strip()
    if 'Ozone' in df.columns:
        df.rename(columns={'Ozone': 'O3'}, inplace=True)
    for col in ['Date', 'DATE', 'Datetime', 'timestamp']:
        if col in df.columns:
            df.rename(columns={col: 'date'}, inplace=True)
    return df

# ü§ñ Enhanced XGBoost Modeling Function

def build_and_evaluate_xgboost(df, feature_cols, target, city_name):
    if 'date' in df.columns:
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        df['month'] = df['date'].dt.month
        df['dayofweek'] = df['date'].dt.dayofweek

    # Feature engineering: lag and rolling stats
    for col in ['SO2', 'PM2.5', 'PM10']:
        if col in df.columns:
            df[f'{col}_lag1'] = df[col].shift(1)
            df[f'{col}_lag2'] = df[col].shift(2)
            df[f'{col}_lag3'] = df[col].shift(3)
            df[f'{col}_lag7'] = df[col].shift(7)
            df[f'{col}_roll3'] = df[col].rolling(window=3).mean()
            df[f'{col}_std3'] = df[col].rolling(window=3).std()

    df = df.dropna()

    engineered_features = feature_cols + ['month', 'dayofweek']
    for col in df.columns:
        if any(key in col for key in ['lag', 'roll', 'std']):
            engineered_features.append(col)

    available_features = list(set(engineered_features) & set(df.columns))
    if target not in df.columns:
        raise ValueError(f"Target {target} not found in dataset: {city_name}")

    print(f"\nüßπ {city_name} - Using features: {available_features}")

    X = df[available_features]
    y = df[target]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    split = int(len(X) * 0.8)
    X_train, X_test = X_scaled[:split], X_scaled[split:]
    y_train, y_test = y.iloc[:split], y.iloc[split:]

    model = XGBRegressor(n_estimators=200, max_depth=6, learning_rate=0.1, objective='reg:squarederror', random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)

    print(f"üìç {city_name} - XGBoost RMSE: {rmse:.2f}, R¬≤: {r2:.2f}, MAPE: {mape*100:.2f}%")

    # Prediction Plot
    plt.figure(figsize=(10, 5))
    plt.plot(y_test.values, label='Actual')
    plt.plot(y_pred, label='Predicted')
    plt.title(f'{city_name} - XGBoost Prediction for {target}')
    plt.legend()
    plt.show()

    # SHAP Analysis
    explainer = shap.Explainer(model)
    shap_values = explainer(X_test)

    shap.summary_plot(shap_values, X_test, feature_names=available_features, plot_type="bar")
    shap.summary_plot(shap_values, X_test, feature_names=available_features)

    return model, rmse, r2, mape

# üéØ Features & Target
feature_cols = ['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'Temp', 'RH', 'WS']
target = 'SO2'

# Run XGBoost for Gaziabad
df_gz = load_dataset('/content/Gaziabad_knn_imputed_data.xlsx')
xgb_model_gz, rmse_gz, r2_gz, mape_gz = build_and_evaluate_xgboost(df_gz, feature_cols, target, 'Gaziabad')

# Run XGBoost for Huairou
df_hr = load_dataset('/content/Huairou_IIET.csv')
xgb_model_hr, rmse_hr, r2_hr, mape_hr = build_and_evaluate_xgboost(df_hr, feature_cols, target, 'Huairou')
```
